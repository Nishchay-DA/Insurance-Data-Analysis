##### Step 1: Import libraries such as Pandas, matplotlib, NumPy, and seaborn and load the insurance dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('insurance.csv')
data
```

##### Step 4: Exploring the Relationship between the Feature and Target Columns

```python
fig,axs=plt.subplots(1,3,sharey=True)
data.plot(kind='scatter',x='age',y='charges',ax=axs[0],figsize=(16,8))
data.plot(kind='scatter', x='bmi', y='charges', ax=axs[1])
data.plot(kind='scatter', x='children', y='charges', ax=axs[2])
```

```python
sns.countplot(data=data,x='region',palette='pastel')
plt.title('Regions')
plt.show()
```

```python
sns.countplot(data=data,x='sex',palette='pastel')
plt.title('Sex')
```

```python
sns.countplot(data=data,x='smoker',palette='pastel')
plt.title('Smoker')
```

#### Step 5: Plotting of Feature vs Feature Plots

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='children',y='age',data=data,palette='pastel')
plt.show()
```

```python
# Bar plot for Age vs Charges
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="age", y="charges", ci=None, palette="viridis")
plt.title("Bar Plot: Age vs Charges", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Average Charges", fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot for Region vs Charges
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="region", y="charges", ci=None, palette="Set2")
plt.title("Bar Plot: Region vs Charges", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Average Charges", fontsize=12)
plt.show()
```

```python
# Bar plot for Sex vs Charges
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x="sex", y="charges", ci=None, palette="coolwarm")
plt.title("Bar Plot: Sex vs Charges", fontsize=14)
plt.xlabel("Sex", fontsize=12)
plt.ylabel("Average Charges", fontsize=12)
plt.show()
```

```python
# Bar plot for BMI vs Charges (Grouping BMI into bins for clarity)
data['bmi_bins'] = pd.cut(data['bmi'], bins=10, labels=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="bmi_bins", y="charges", ci=None, palette="magma")
plt.title("Bar Plot: BMI vs Charges", fontsize=14)
plt.xlabel("BMI Bins", fontsize=12)
plt.ylabel("Average Charges", fontsize=12)
plt.show()
```

```python
# Bar plot for Smoker vs Charges
plt.figure(figsize=(6, 4))
sns.barplot(data=data, x="smoker", y="charges", ci=None, palette="rocket")
plt.title("Bar Plot: Smoker vs Charges", fontsize=14)
plt.xlabel("Smoker", fontsize=12)
plt.ylabel("Average Charges", fontsize=12)
plt.show()
```

#### Step 6: Plotting the age vs charges graph

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Scatter Plot: Age vs Charges with Smoker Consideration
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="age", y="charges", hue="smoker", palette="coolwarm", alpha=0.8)
plt.title("Scatter Plot: Age vs Charges (Considering Smoker)", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Charges", fontsize=12)
plt.legend(title="Smoker", loc="upper left")
plt.show()
```
